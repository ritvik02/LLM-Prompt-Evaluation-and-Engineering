```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
cko = pd.read_excel('cks.xlsx')
cko
```

```python
# cko is data frame you get from excel



#code for extracting transcripts

for i, row in cko.iterrows():

    url = row['Transcript Link']

    if pd.isna(url):

        continue

    else:

        res = requests.get(url)

        html_page = res.content

        soup = BeautifulSoup(html_page, 'html.parser')

        text = soup.find_all(text=True)

        transcript = text[0]

        cko.loc[i,'Transcript text'] = str(transcript)



```

```python
cko['Transcript text'][0]
```

```python
cko.to_excel('cko_new.xlsx', index=False)
```

```python
#code for extracting summaries

cko['summary_exists'] = [True if 'SUMMARY' in str(text) else False for text in cko['Transcript text'] ]
cko_summary = cko[cko['summary_exists'] == True]
cko_summary['summary'] = [summary.split('SUMMARY')[1] for summary in cko_summary['Transcript text']]
```

```python
cko.to_excel('cko_summary.xlsx', index=False)
```
